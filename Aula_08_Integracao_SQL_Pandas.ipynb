```python
import pandas as pd
import sqlite3
import os

# 1. Conectar no Banco (ou criar se sumiu)
conn = sqlite3.connect('loja_big_tech.db')
cursor = conn.cursor()

# 2. Garantir que a tabela existe e est√° vazia (Reset Limpo)
cursor.execute("DROP TABLE IF EXISTS produtos")
cursor.execute("""
CREATE TABLE produtos (
    id_produto INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT,
    categoria TEXT,
    preco REAL,
    estoque INTEGER
)
""")
conn.commit()

print("‚úÖ Base de Dados Pronta e Conectada!")

# Criando um DataFrame (Simulando o Excel que chegou do chefe)
dados_novos = {
    'nome': ['Galaxy S24', 'Monitor Dell', 'Teclado HyperX', 'Mouse Gamer', 'Alexa Echo', 'PS5 Slim', 'Cadeira Ergonomica'],
    'categoria': ['Celulares', 'Inform√°tica', 'Acess√≥rios', 'Acess√≥rios', 'Smart Home', 'Games', 'M√≥veis'],
    'preco': [5500.00, 1200.50, 450.00, 120.00, 350.00, 3800.00, 1500.00],
    'estoque': [100, 40, 50, 60, 200, 15, 8]
}

df_excel = pd.DataFrame(dados_novos)

print("--- O ARQUIVO DO CHEFE (NO PANDAS) ---")
display(df_excel)

# O COMANDO M√ÅGICO
# nome da tabela, conex√£o, if_exists='append' (adicionar ao que j√° tem), index=False (n√£o enviar o n√∫mero da linha 0,1,2...)
df_excel.to_sql('produtos', conn, if_exists='append', index=False)

print("üöÄ Sucesso! 7 produtos inseridos em milissegundos.")

# Verificando o banco
df_conferencia = pd.read_sql("SELECT * FROM produtos", conn)
display(df_conferencia)

# A M√ÅGICA DA AGREGA√á√ÉO (SUM, AVG, COUNT)
# Estamos pedindo 3 coisas numa tacada s√≥:
# 1. Quantos tipos de produtos temos? (COUNT)
# 2. Qual a m√©dia de pre√ßo? (AVG)
# 3. A GRANA: Pre√ßo * Estoque somado (SUM)

query_auditoria = """
SELECT
    COUNT(*) as tipos_produtos,
    AVG(preco) as preco_medio,
    SUM(preco * estoque) as valor_total_estoque
FROM produtos
"""

df_financeiro = pd.read_sql(query_auditoria, conn)

print("--- üìä RELAT√ìRIO FINANCEIRO DA LOJA ---")
display(df_financeiro)
```
